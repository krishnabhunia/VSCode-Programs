```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.array(plt.imread("./images/robo.jpeg"))
plt.imshow(image)
plt.show()
```

```python
image_r  = image.copy()
image_g  = image.copy()
image_b  = image.copy()
image_r[:,:,(1,2)]=0
image_g[:,:,(0,2)]=0
image_b[:,:,(0,1)]=0
# image_r[:,:,:] accesses all rows and columns (: means “all”), and (1, 2) refers to the second and third color 
# channels (1 for green, 2 for blue). By assigning 0 to these channels, it leaves only the red channel intact.
plt.imshow(np.hstack((image_r,image_g,image_b)))
plt.show()
```

```python
image_r  = image.copy()
image_g  = image.copy()
image_b  = image.copy()
image_r[:,:,(1,2)]=100
image_g[:,:,(0,2)]=100
image_b[:,:,(0,1)]=100
plt.imshow(np.hstack((image_r,image_g,image_b)))
plt.show()
```
